A particle is in the potential

V(x) = (e ^ (-2x))/x + 3x .

If the total energy of the particle is 15 units, find the range of its motion. Also find the equilibrium position.

- fsolve: is used to find the roots of an equation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

```python
# Define the potential function
def V(x):
    return (np.exp(-2 * x) / x) + 3 * x

# Define the derivative of the potential function
def dV_dx(x):
    return (-2 * np.exp(-2 * x) / x - np.exp(-2 * x) / (x ** 2)) + 3
```

```python
# Find the equilibrium position (where dV/dx = 0)
equilibrium_position = fsolve(dV_dx, 1)[0] # Initial guess is 1

# Find the range of motion (where V(x) = 15)
def energy_equation(x):
    return V(x) - 15

# Find the points where V(x) = 15
x_range = fsolve(energy_equation, [0.1, 5]) # Initial guesses

# Print results
print(f"Equilibrium Position: {equilibrium_position}")
print(f"Range of Motion: {x_range}")
```

```python
# Optional: Plot the potential energy
x_values = np.linspace(0.1, 5, 400)
V_values = V(x_values)

plt.plot(x_values, V_values, label='Potential V(x)')
plt.axhline(y=15, color='r', linestyle='--', label='Total Energy (E=15)')
plt.scatter(x_range, [15, 15], color='black') # Mark the range of motion
plt.title('Potential Energy vs Position')
plt.xlabel('Position (x)')
plt.ylabel('Potential Energy V(x)')
plt.legend()
plt.grid()
plt.show()
```
