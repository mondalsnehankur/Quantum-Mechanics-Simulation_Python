Quantum Harmonic Oscillator


importing required python libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint,simps
from scipy.optimize import newton,bisect
```

plotting energies

```python
def f(u,x,E):
  f1,f2=u[1],(x**2-2*E)*u[0]
  return(f1,f2)
def shoot(E):
  sol=odeint(f,u,x,args=(E,))
  return sol[:,0][-1]
energies=np.arange(0,6,0.1)
x=np.linspace(-5,5,100)
u=[0,0.001]
hits=[shoot(E) for E in energies]
plt.ylim(-2500,2500)
plt.axhline()

plt.plot(energies,hits)
```

normalized wave function

```python
def psi_n(En):
  sol=odeint(f,u,x,args=(En,))
  psi=sol[:,0]
  N=1/np.sqrt(simps(psi*psi,x))
  return N*psi
```

plotting wave functions

```python
#ground state
En=bisect(shoot,0,1)
plt.plot(x,psi_n(En))


#second state
En=bisect(shoot,1,2)
plt.plot(x,psi_n(En))

#third state
En=bisect(shoot,2,3)
plt.plot(x,psi_n(En))

#fourth state
En=bisect(shoot,3,4)
plt.plot(x,psi_n(En))
```

plotting probability densities


```python
#ground state
En=bisect(shoot,0,1)
plt.plot(x,(psi_n(En))**2)

#second state
En=bisect(shoot,1,2)
plt.plot(x,(psi_n(En))**2)

#third state
En=bisect(shoot,2,3)
plt.plot(x,(psi_n(En))**2)

#fourth state
En=bisect(shoot,3,4)
plt.plot(x,(psi_n(En))**2)

```

printing eigen energies


```python
#ground state
E0=bisect(shoot,0,1)
print(E0)
#first excited state/ first state
E1=bisect(shoot,1,2)
print(E1)
#second excited state/third state
E2=bisect(shoot,2,3)
print(E2)
```

plotting E2(1+0.02) and E2(1-0.02)


```python
plt.plot(x,psi_n(E2*(1.0+0.02)),label='plus')
plt.plot(x,psi_n(E2*(1.0-0.02)),label='minus')
```

orthogonality


```python
En0=bisect(shoot,0,1)
psi0=psi_n(En0)
En1=bisect(shoot,1,2)
psi1=psi_n(En1)

orth=simps(psi1*psi1,x)
print(orth)




```
